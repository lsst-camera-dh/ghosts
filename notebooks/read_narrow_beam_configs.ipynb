```python
def parse_hdr_kvals(file_path):
    """ parse AR CCOB NB configurations
    """
    content = open(file_path, "r").readlines()
    nb_config_list = []
    for i in range(len(content)):
        one = content[i]
        # RawData/13579/MC_C_20231121_000001/MC_C_20231121_000001_R22_S11.fits
        if one.find('fits')>0:
            if i>0:
                nb_config_list.append(nb_config)
            nb_config = {}
            run_num = one.split('/')[-3]
            day_obs = one.split("_")[2]
            seq_num = one.split("_")[3].strip("/MC")
            #print(day_obs, seq_num)
            nb_config["run"] = int(run_num)
            nb_config["day_obs"] = day_obs
            nb_config["seq_num"] = seq_num
        elif one.find('=')>0:
            tok = one.split("=")[0]
            val = one.split("=")[1].split('\t')[0]
            comment = one.split("=")[1].split('\t')[2]
            #print(tok, val, comment)
            nb_config[tok.strip()] = val
            nb_config[tok.strip()+'_Comment'] = comment
    return nb_config_list

def build_config_df(list_dict):
    """ Make Data Frame from a list of config dicts
    """
    df = pd.DataFrame(list_dict)
    df['Camera_X'] = pd.to_numeric(df['CATKW105'])
    df['Camera_Y'] = pd.to_numeric(df['CATKW107'])
    df['Camera_Z'] = 3.395711882*1000 - pd.to_numeric(df['CATKW109']) # 3.395712 - 0.327662 (carriage-L1S1 plane sep in mm)
    df['Camera_U'] = pd.to_numeric(df['CATKW103'])
    df['Camera_B'] = pd.to_numeric(df['CATKW101'])
    df.head()
    return df

def process_configs(file_path):
    """ Read, parse and make a data frame
    """
    return build_config_df(parse_hdr_kvals(file_path))

def getXYZUB(df, run=13579, day_obs="20231121", seq_num="000001"):
    """ Helper to get useful quantities
    """
    # check that a configuration is found
    if df[ (df['run']==run) & (df['day_obs']==day_obs) & (df['seq_num']==seq_num) ].Camera_X.size>0:        
        X = df[ (df['run']==run) & (df['day_obs']==day_obs) & (df['seq_num']==seq_num) ].Camera_X.item()
        Y = df[ (df['run']==run) & (df['day_obs']==day_obs) & (df['seq_num']==seq_num) ].Camera_Y.item()
        Z = df[ (df['run']==run) & (df['day_obs']==day_obs) & (df['seq_num']==seq_num) ].Camera_Z.item()
        U = df[ (df['run']==run) & (df['day_obs']==day_obs) & (df['seq_num']==seq_num) ].Camera_U.item()
        B = df[ (df['run']==run) & (df['day_obs']==day_obs) & (df['seq_num']==seq_num) ].Camera_B.item()
        return X, Y, Z, U ,B
    else:
        print("No configuration found for:", run, day_obs, seq_num)
        return None
```

```python
import copy
from ghosts.beam import rotate_config_to_ub, concat_dicts
from ghosts.beam_configs import BEAM_CONFIG_0, BEAM_CONFIG_1, BEAM_CONFIG_3

```

```python
def beam_config_from_data(df, run_id=13579, day_obs="20231121", seq_num="000001"):
    """
    """
    x, y, z, u, b = getXYZUB(df, run_id, day_obs, seq_num)
    print(f'x={x}, y={y}, z={z}, u={u}, b={b}')
    new_beam = copy.deepcopy(BEAM_CONFIG_0)
    new_beam['beam_id'] = day_obs+seq_num
    new_beam['n_photons'] = 100
    new_beam['x_offset'] = y/1000. # 86.561/1000.   # 0.012 # 0.022 # -0.012  # -6.5/1000.
    new_beam['y_offset'] = x/1000. # 57.653/1000. # 0.114 # 0.26  #  0.114  #  100/1000.
    new_beam['z_offset'] = z/1000.  # 3.06
    new_beam = rotate_config_to_ub(new_beam, u, b)
    return new_beam
    
```
